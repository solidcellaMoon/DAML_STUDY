원래 인덱스랑 바뀐 후 인덱스 차이가 2이상이면 불가능한거?

```python
from collections import deque

def minimumBribes(q):
    origin = deque([i for i in range(1,len(q)+1)]) #원상태
    bribes = [0 for _ in range(len(q))] #각 사람의 뇌물 수
    cnt = 0
    
    while origin != q:
        
        target = origin.pop()
        
        bribeNum = abs(q.index(target) - (target-1))
        
        if bribeNum > 2:
            return 'Too chaotic'
        elif bribeNum != 0:
            origin.insert(q.index(target), target)
            bribes[target-1] = bribeNum
            
        print(origin)
        print(bribes)

    return sum(bribes)
    
minimumBribes([2,1,5,3,4])
```
