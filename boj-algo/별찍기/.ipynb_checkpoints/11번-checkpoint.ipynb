풀이: https://jinho-study.tistory.com/88

```python
import math

s = ["  *   ", " * *  ", "***** "]

def stars(shift):
    for i in range(len(s)):
        print('-'*70)
        print('append',s[i]+s[i])
        print('-'*70)
        s.append(s[i]+s[i])
        s[i] = (' '*3*shift + s[i] + ' '*3*shift )
        print('\n'.join([''.join([str(i) for i in row]) for row in s ]))
        
n = int(input())
k = int(math.log(int(n/3),2))

for i in range(k):
    print('='*30,i+1,'번째','='*30)
    stars(int(pow(2,i)))
    
print('결과','-'*70)
for i in range(n):
    print(s[i])
```
